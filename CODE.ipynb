```python
# IMPORTING THE LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
# LOADING THE DATASET 
DATAPATH = 'mushrooms.csv'
data = pd.read_csv(DATAPATH)
data.head(8124)
```

```python
# PERFORMING EXPOLATORY DATA ANALYSIS (EDA)

# Checking whether the dataset is balance or not
x = data['class']
ax = sns.countplot(x=x,data=data,palette="Set3")
```

```python
# Analysis with respect to all 23 attributes by using helper function
def plot_data(hue,data):
    for i,col in enumerate(data.columns):
        plt.figure(i)
        sns.set(rc={'figure.figsize':(11.7,8.27)})
        ax = sns.countplot(x=data[col], hue=hue, data=data,palette="Set3")
hue = data['class']
data_for_plot = data.drop('class', 1)

plot_data(hue, data_for_plot)
```

```python
# Checking whether there are any missing columns
for col in data.columns:
    print("{} : {}".format(col, data[col].isnull().sum()))
```

```python
# LABLE ENCODING AND ONE_HOT ENCODING

# Lable encoding on target column
le = LabelEncoder()
data['class'] = le.fit_transform(data['class'])
data.head(8124)
```

```python
# One-hot encoding on target column
encoded_data = pd.get_dummies(data)
encoded_data.head(8124)
```

```python
# SEPARATING LABELS FROM THE FEATURES
y = encoded_data['class'].to_frame()
X = encoded_data.drop('class', axis=1)
```

```python
y.head(8124)
```

```python
X.head(8124)
```

```python
# STANDARDISING OUR FEATURES

# Importing library for standardising
from sklearn.preprocessing import StandardScaler

# Standardising the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# CREATING TRAINING AND TESTING DATASET

# Importing library for seprating training and testing data
from sklearn.model_selection import train_test_split

# Data seprated as 70% training data set and 30% as testing data set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, random_state=19)
```

```python
# LOGISTIC REGRESSION

# Importing libraries for logistic regression and confusion matrix
from sklearn.model_selection import cross_val_score

#from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Performing logistic regression
# Creating variable
logistic_reg = LogisticRegression()

# Fiting of plot
logistic_reg.fit(X_train, y_train.values.ravel())

y_pred_test = logistic_reg.predict(X_test)

# Calculating the accuracy
print('Accuracy of Logistic Regression classifier on the test set: {:.2f}'.format(accuracy_score(y_test, y_pred_test)))
```

```python
# CONFUSION MATRIX
from sklearn import metrics
confusion_matrix=metrics.confusion_matrix(y_test,y_pred_test)
confusion_matrix
```

```python
# Ploting confusion matrix
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]

f,ax = plt.subplots(figsize =(8,8))
sns.heatmap(confusion_matrix,cmap=plt.cm.Blues)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title('CONFUSION MATRIX FOR LOGISTIC REGRESSION')
plt.show()
```

```python
# PLOTING RECEIVER OPERATING CHARACTERISTIC CURVE

# Importing libraries for ploting roc curve
from sklearn.metrics import roc_curve, auc

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred_test)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python
# VISUALIZE THE ROC CURVE
plt.figure(figsize=(8,8))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='yellow',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```
